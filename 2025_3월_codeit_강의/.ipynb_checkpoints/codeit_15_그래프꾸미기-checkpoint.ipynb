## 그래프 = 소통 수단
처음 보는 사람들도 이해할 수 있게끔 그래프를 적절히 표현해주어야 함

## 그래프 모양 변경 정리
1. 그래프 타이틀 : **plt.title('그래프를 설명할 제목')**
2. x, y축 이름 정하기 : **plt.xlabel('x축 설명'), plt.ylabel('y축 설명')**
3. 그래프 색 변경하기 : plt.그래프모양(x축 데이터, y축 데이터, **c = '원하는 색 코드/색 이름'**)
4. 그래프 모양 변경하기 : plt.그래프모양(x축 데이터, y축 데이터, **marker = '원하는 모양이름/원하는 기호'**)
5. 그래프 크기 변경하기 : **plt.figure(figsize = (x, y))** x = 가로 길이/y = 세로 길이
6. 전체 그래프 크기 통일 시키기 : **plt.rcParams['figure.figsize'] = (x, y)** x = 가로 길이/y = 세로 길이

```python
import numpy as np
import matplotlib.pyplot as plt # matplotlib.pyplot
```

```python
height_array = np.array([
    165, 164, 155, 151, 157, 162, 155, 157, 165, 162,
    165, 167, 167, 183, 180, 184, 177, 178, 175, 181,
    172, 173, 169, 172, 177, 178, 185, 186, 190, 187
])

weight_array = np.array([
    62, 59, 57, 55, 60, 58, 51, 56, 68, 64,
    57, 58, 64, 79, 73, 76, 61, 65, 83, 80,
    67, 82, 88, 62, 61, 79, 81, 68, 83, 80
])
```

```python
# 산점도 그래프를 보기 쉽게 바꾸는 연습을 해보자
plt.scatter(height_array, weight_array)
plt.show()
```

```python
plt.scatter(height_array, weight_array)
plt.title('Height and Weight')     # .title 을 사용하여 그래프의 제목을 알려줌
plt.xlabel('Height (cm)')          # .xlabel x축의 이름을 정해줌
plt.ylabel('Weight (kg)')          # .ylabel y축에 이름을 정해줌
plt.show()
```

```python
plt.scatter(height_array, weight_array, c = 'green') # c = '원하는 색' 으로 그래프 색을 변경할 수 있음 컬러 핵스코드도 사용 가능
plt.title('Height and Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
plt.scatter(height_array, weight_array, c = 'green', marker = '+') # marker = '원하는 모양' 으로 설정 시 그래프 모양이 변경
plt.title('Height and Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
plt.scatter(height_array, weight_array, c = 'green', marker = 's') # 위 그래프 모양을 사각형으로 바꾸고 싶을 땐 square 줄임말 's'를 사용
plt.title('Height and Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

Matplotlib에서는 그래프의 크기를 기본적으로 인치 단위로 표현한다.\
따로 설정하지 않으면 가로 6인치 세로 4인치 크기의 그래프가 나옴.

만약 다른 크기의 그래프를 원할 때는 pyplot의 **figure()** 함수를 호출해서\
**figsize = (x, y)** x = 가로 길이, y = 세로 길이 를 정해주면 된다.

```python
plt.figure(figsize = (10, 4))  # 그래프 크기 변경
plt.scatter(height_array, weight_array, c = 'green', marker = 's')
plt.title('Height and Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

그래프를 그릴 때 마다 일일이 크기를 지정하지 않고 전체적으로 그래프 크기를 통일하고 싶을 수도 있다.\
이럴 때에는 pyplot의 **rcParams** 속성에 직접 접근하여 그래프 사이즈의 기본 설정을 변경할 수 있다.

```python
plt.rcParams['figure.figsize'] = (5, 5) # 전체 그래프 크기 통일 시키기 (따로 그래프 크기를 설정하지 않아도 계속 이 크기로 출력됨)
plt.scatter(height_array, weight_array, c = 'green', marker = 's')
plt.title('Height and Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```
