```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
year_array = np.array([2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]) # 2011 ~ 2020년

stock_array = np.array([
    14.46, 19.01, 20.04, 27.59, 26.32,
    28.96, 42.31, 39.44, 73.41, 132.69
])                                        # 애플 2011 ~ 2020 주가
```

```python
plt.plot(year_array, stock_array)  # matplotlib.pyplot을 사용하여 (x,y)축에 들어갈 값을 .plot 명령어로 넣어줌(연도, 주식 가격)
# plt.show() # 그래프로 출력해달라는 뜻 (하지만 주피터노트북에선 이 명령어를 사용 안해도 사용 가능)
```

```python
# 선 그래프(.plot)
```

```python
# 반장 선거 관련 데이터
name_array = np.array(['mark', 'dongwook', 'hyojune', 'sowon', 'taeho']) # 이름
vote_array = np.array([5, 10, 6, 8, 3])   # 투표 수
```

```python
plt.bar(name_array, vote_array)
plt.show()
```

```python
# 막대 그래프(.bar)
```

```python
# 사람들의 키 데이터
height_array = np.array([
    165, 164, 155, 151, 157, 162, 155, 157, 165, 162,
    165, 167, 167, 183, 180, 184, 177, 178, 175, 181,
    172, 173, 169, 172, 177, 178, 185, 186, 190, 187
])

# 사람들의 몸무계 데이터
weight_array = np.array([
    62, 59, 57, 55, 60, 58, 51, 56, 68, 64,
    57, 58, 64, 79, 73, 76, 61, 65, 83, 80,
    67, 82, 88, 62, 61, 79, 81, 68, 83, 80
])
```

```python
plt.scatter(height_array, weight_array)
plt.show()
```

```python
# 산점도 그래프(.scatter) 두 변수의 연관성을 파악할 때 사용, 키와 몸무계가 서로 어떻게 연관되어있는지 보는 것
# 키가 작으면 몸무계도 적고 키가 크면 몸무계도 많이 나감. (= 두 변수(몸무계, 키)는 상관 관계를 갖고 있다.)
```
