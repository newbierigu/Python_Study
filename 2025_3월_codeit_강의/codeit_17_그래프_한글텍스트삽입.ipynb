## 지금까지 그래프에는 영어로된 설명이 들어갔지만, 이번엔 한글로 표현해보자.

```python
import numpy as np
import matplotlib.pyplot as plt

height_array = np.array([
    165, 164, 155, 151, 157, 162, 155, 157, 165, 162,
    165, 167, 167, 183, 180, 184, 177, 178, 175, 181,
    172, 173, 169, 172, 177, 178, 185, 186, 190, 187
])

weight_array = np.array([
    62, 59, 57, 55, 60, 58, 51, 56, 68, 64,
    57, 58, 64, 79, 73, 76, 61, 65, 83, 80,
    67, 82, 88, 62, 61, 79, 81, 68, 83, 80
])
```

```python
plt.scatter(height_array, weight_array, c= 'purple', marker= 's')
plt.title('키와 몸무계')
plt.xlabel('키 (cm)')
plt.ylabel('몸무계 (kg)')
plt.show()
```

원래 제목과 키, 몸무계의 **이름을 넣는 곳에 한글을 입력 시 이런식으로 에러가 발생하며 그래프의 한글의 깨진 상태로 출력된다.**

이 문제를 해결하기 위해서 **폰트를 바꿔주는 코드**를 사용해야 한다.

```python
plt.scatter(height_array, weight_array, c= 'purple', marker= 's')
plt.rc('font', family= 'Malgean Gothic')
plt.title('키와 몸무계')
plt.xlabel('키 (cm)')
plt.ylabel('몸무계 (kg)')
plt.show()
```

### Window 환경에서는 Malgean Gothic(글씨체 맑은 고딕) 이 기본적으로 내장되어있으므로 해당 폰트를 사용.\
