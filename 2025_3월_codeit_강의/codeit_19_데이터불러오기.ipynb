## csv에 column명이 없으면 따로 설정해주면 된다.
burger_df_2 = pd.read_csv("data/burger2.csv", header=None,\
**names = 리스트** 를 사용하여 각 column 마다 이름을 순서대로 붙여주면 된다.\
**names = ["product_name", "calories", "carb,protein", "fat,sodium", "category"])**\
burger_df_2

```python
burger_df_2 = pd.read_csv("data/burger2.csv", header=None,
                          names=["product_name", "calories", "carb", "protein", "fat", "sodium", "category"])
burger_df_2
```

위 표의 왼쪽 0, 1, 2 ...들은 각 rows의 인덱스이다.\
각 columns 들에게 이름을 붙여줬듯이 **rows들에게도 이름을 붙여보자.**\
burger_df = pd.read_csv("data/burger.csv", **index_col="product_name"**)\
**index_col=""** 은 각 index들을 **"해당 값"** 으로 변경해주겠다는 코드이다.\
참고로 인덱스는 절대로 겹치면 안됨 겹치면 오류발생
또 index_col에다 column들의 이름 대신 인덱스 번호로도 불러올 수 있음(헤더 미포함)\
**index_col="product_name"은 == index_col=0** 이다.

```python
burger_df = pd.read_csv("data/burger.csv", index_col=0)
burger_df
```
